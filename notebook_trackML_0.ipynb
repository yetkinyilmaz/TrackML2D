```python
from sklearn.cluster import DBSCAN
```

```python
def polar(xyz):
    ptsnew = np.zeros(xyz.shape)
    xy = xyz[:,0]**2 + xyz[:,1]**2
    ptsnew[:,0] = np.sqrt(xy)/10000000
    ptsnew[:,1] = np.arctan2(xyz[:,1], xyz[:,0])
    return ptsnew
```
